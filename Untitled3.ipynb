<p style="line-height:1.55;">

earthquake-triggered landslides. Using pgmpy, the model learns CPDs from a simulated dataset
and performs inference through Variable Elimination. Key outputs include:

</ul>

# Infered pre extracted data from the paper

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import numpy as np
import pandas as pd


#from the paper’s Fig. 2

model = DiscreteBayesianNetwork([
    ('Magnitude', 'PGA'),
    ('Distance', 'PGA'),
    ('PGA', 'Landslide'),
    ('Landslide', 'BuildingDamage'),
    ('Decision', 'Measure'),
    ('Decision', 'Alarm'),
    ('Decision', 'Cost'),
    ('Measure', 'Landslide'),
    ('BuildingDamage', 'Utility')
])


#define node states

mag_states = ['Low', 'High']
dist_states = ['Near', 'Far']
pga_states = ['Low', 'High']
land_states = ['No', 'Yes']
bld_states = ['NoDamage', 'SomeDamage', 'Collapse']
decision_states = ['Passive', 'Active', 'NoAction', 'WarningSystem']
measure_states = ['No', 'Yes']
alarm_states = ['No', 'Yes']
cost_states = ['Low', 'Medium', 'High']
utility_states = ['Good', 'Moderate', 'Poor']

#CPDs
#magnitude (prior)
cpd_mag = TabularCPD('Magnitude', 2, [[0.7], [0.3]])

#distance (prior)
cpd_dist = TabularCPD('Distance', 2, [[0.6], [0.4]])

# PGA | Magnitude, Distance  (strong = highmag + near)
cpd_pga = TabularCPD('PGA', 2,
    [[0.9, 0.6, 0.5, 0.2],  #low
     [0.1, 0.4, 0.5, 0.8]], #high
    evidence=['Magnitude', 'Distance'], evidence_card=[2, 2])

# Landslide | PGA, Measure  (from table 3 and 4)
cpd_landslide = TabularCPD('Landslide', 2,
    [[0.85, 0.93, 0.4, 0.7],   # no Landslide
     [0.15, 0.07, 0.6, 0.3]],  # landslide
    evidence=['PGA', 'Measure'], evidence_card=[2, 2])

# Building Damage | Landslide  (table5)
cpd_build = TabularCPD('BuildingDamage', 3,
    [[0.8, 0.3],   #no damage
     [0.15, 0.4],  #some damage
     [0.05, 0.3]], #collapse
    evidence=['Landslide'], evidence_card=[2])

# Decision (prior, equal probability)
cpd_decision = TabularCPD('Decision', 4, np.ones((4,1))/4)

# Measure | Decision (table 6: WarningSystem  Yes)
cpd_measure = TabularCPD('Measure', 2,
    [[0.9, 0.7, 1.0, 0.1],  # No
     [0.1, 0.3, 0.0, 0.9]], # Yes
    evidence=['Decision'], evidence_card=[4])


# Alarm | Decision (Table 7: WarningSystem gives reliable alarm)
cpd_alarm = TabularCPD('Alarm', 2,
    [[0.7, 0.5, 0.6, 0.1],  # No
     [0.3, 0.5, 0.4, 0.9]], # Yes
    evidence=['Decision'], evidence_card=[4])

# Cost | Decision (Table 9)
cpd_cost = TabularCPD('Cost', 3,
    [[0.6, 0.2, 0.8, 0.1],  # Low
     [0.3, 0.5, 0.2, 0.4],  # Medium
     [0.1, 0.3, 0.0, 0.5]], # High
    evidence=['Decision'], evidence_card=[4])

# Utility | BuildingDamage (Table 10)
cpd_util = TabularCPD('Utility', 3,
    [[0.9, 0.1, 0.0],   #good
     [0.1, 0.8, 0.2],   #moderate
     [0.0, 0.1, 0.8]],  #poor
    evidence=['BuildingDamage'], evidence_card=[3])


# 4. Add CPDs to model and check

model.add_cpds(cpd_mag, cpd_dist, cpd_pga, cpd_landslide, cpd_build,
               cpd_decision, cpd_measure, cpd_alarm, cpd_cost, cpd_util)
assert model.check_model()
print("BN is validated.\n")




infer = VariableElimination(model)
#1 strong nearby earth quake with close distance
res1 = infer.query(['BuildingDamage'], evidence={'Magnitude':1, 'Distance':0})
print("P(BuildingDamage | High Magnitude, Near Distance):")
print(res1)


utility_map = {'Good': 0, 'Moderate': -10000, 'Poor': -20000}
utility_state_names = ['Good', 'Moderate', 'Poor']

def expected_utility_for_decision(decision_idx):
    q = infer.query(['Utility'], evidence={'Decision': decision_idx})
    exp_util = sum(q.values[i] * utility_map[utility_state_names[i]] for i in range(3))
    return exp_util

decisions = ['Passive', 'Active', 'NoAction', 'WarningSystem']
results = [(d, expected_utility_for_decision(i)) for i, d in enumerate(decisions)]
df = pd.DataFrame(results, columns=['Decision', 'Expected_Utility'])
print("\nExpected Utilities (higher is better):\n", df)
```

#Trained using simulated dataset

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd
import numpy as np

# load and preprocess
df = pd.read_csv('simulated_earthquake_landslide_dataset.csv')
df.columns = df.columns.str.strip()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()

print("DS loaded Shape:", df.shape)

model = DiscreteBayesianNetwork([
    ('Magnitude', 'PGA'),
    ('Distance', 'PGA'),
    ('PGA', 'Landslide'),
    ('Landslide', 'BuildingDamage'),
    ('Decision', 'Measure'),
    ('Decision', 'Alarm'),
    ('Decision', 'Cost'),
    ('Measure', 'Landslide'),
    ('Decision', 'Utility')
])

# learn CPD parameters
model.fit(df, estimator=MaximumLikelihoodEstimator)
model.check_model()

print("Model training complete. CPDs learned successfully.\n")

# inference model
infer = VariableElimination(model)

# inference examples
print("P(BuildingDamage | Magnitude=High, Distance=Near):")
print(infer.query(['BuildingDamage'],
                  evidence={'Magnitude': 'High', 'Distance': 'Near'}), "\n")

print("P(Landslide | PGA=High):")
print(infer.query(['Landslide'], evidence={'PGA': 'High'}), "\n")

print("P(Landslide | PGA=High, Measure=Yes):")
print(infer.query(['Landslide'], evidence={'PGA': 'High', 'Measure': 'Yes'}), "\n")

# EU mapping
utility_map = {'Good': 0, 'Moderate': -10000, 'Poor': -20000}
utility_states = ['Good', 'Moderate', 'Poor']

def expected_utility(decision):
    q = infer.query(['Utility'], evidence={'Decision': decision})
    return sum(q.values[i] * utility_map[utility_states[i]]
               for i in range(len(utility_states)))

decisions = df['Decision'].unique()
results = [(d, expected_utility(d)) for d in decisions]

results_sorted = sorted(results, key=lambda x: x[1], reverse=True)
best_dec, best_util = results_sorted[0]

print("Expected Utility by Decision:")
for d, eu in results_sorted:
    print(f"  {d}: {eu:.2f}")

print(f"\nOptimal Decision: {best_dec} (EU = {best_util:.2f})")
```
